<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
# @@@ 8

# X =
# y = 
# 예측에 사용할 변수들 (X)
# 주의: 여기서도 dropna()를 해줘야 X 안에 빈칸(NaN)이 없어집니다.
X = df[['ENROLL_8', 'AVG_MATH_4_SCORE', 'AVG_READING_4_SCORE', 'INSTRUCTION_EXPENDITURE_PER_STUDENT']].dropna()

# 정답 (y)
# X의 인덱스에 맞춰서 y를 가져와야 개수가 맞습니다.
y = df.loc[X.index]['AVG_MATH_8_SCORE']
```

```python
# @@@ 9 

# X_train, X_test, y_train, y_test = train_test_split(
#      X, y, test_size=, random_state=42)

# Tutorial 1처럼 7:3 비율로 나눔
# 0을 채우거나 median으로 채우는 건 선택사항이지만, dropna()를 위에서 했으면 굳이 안 해도 됨.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# @@@ 11
# create your model here
# model = 
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
# @@@ 12
# R^2 Score (1에 가까울수록 좋음)
print("R^2 Score:", model.score(X_test, y_test))

# RMSE (Root Mean Squared Error) - 평균적으로 몇 점 정도 틀리는지
print("RMSE:", np.mean((model.predict(X_test)-y_test)**2)**0.5)
```
